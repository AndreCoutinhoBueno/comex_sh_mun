# Long

Fertilizante

Carrega, elimina ano incompleto, agrupa em na unidade de tempo proposta para o estudo, oredena.

```python
i=pd.read_csv('~/git/comex_sh_mun/temp/i_f_m_r_pa_clean.csv')
i=i[i.CO_ANO!=2024]
i=i.groupby(by=['CO_ANO'])[['KG_f','VL_f']].sum().reset_index()
i.sort_values(by=['CO_ANO'])
```

Adianta 1 unidade de tempo as variáveis quye serão utilizadas como input,elimina menor ano, que ficou sem input.

```python
i1=i[['CO_ANO','KG_f']]
i2=i.drop(columns='KG_f')
i2.CO_ANO=i2.CO_ANO+1
i=i1.merge(i2,how='outer')
i.to_csv('~/git/comex_sh_mun/temp/i.csv',index=False)
i
```

Café

Carrega, adinata uma unidade de tempo, já que todas as variáveis desta tabela serão utilizadas como input. Elimina ano incompleto, agrupa em na unidade de tempo proposta para o estudo, oredena.

```python
e=pd.read_csv('~/git/comex_sh_mun/temp/e_c_m_r_pa_clean.csv')
e=e[e.CO_ANO!=2024]
e=e.groupby(by=['CO_ANO'])[['KG_c','VL_c']].sum().reset_index()
e.CO_ANO=e.CO_ANO+1
e
```

Funde i + e

```python
i_e=i.merge(e,how='outer')
i_e.to_csv('~/git/comex_sh_mun/temp/i_e.csv',index=False)
i_e
```

Chuvas

```python
c=pd.read_csv('~/git/comex_sh_mun/temp/chuvas_m_r_pa_clean.csv')
c=c[c.CO_ANO!=2024]

c=c.groupby(by=['CO_ANO','Estação']).mm.sum().reset_index()
c.CO_ANO=c.CO_ANO+1
c=c.pivot(index='CO_ANO',columns='Estação',values='mm')
c=c.reset_index()
c
```

Funde fertilizantes, café e chuvas com chuvas

```python
i_e_c=i_e.merge(c)
i_e_c.to_csv('~/git/comex_sh_mun/temp/i_e_c.csv',index=False)
i_e_c
```

# WIDE

Fertilizante

```python
i_wide=pd.read_csv('~/git/comex_sh_mun/temp/i_f_m_r_pa_clean_wide.csv')
i_wide
```

O ano de 2025 aparece na tabela por adiantamento de dado incomplto de 2024 presente na base de dados e por isso será eliminado. O valor da variável target, KG_f, aparece no ano de 2024. porém, como o ano esta em andamento, esta incolpeto e também será eliminado.

Salva.

```python
i_wide=i_wide[i_wide.CO_ANO!=i_wide.CO_ANO.max()]
i_wide.loc[i_wide.CO_ANO==2024,'KG_f']=np.nan
i_wide.to_csv('~/git/comex_sh_mun/temp/i_wide.csv',index=False)
i_wide
```

Café

```python
e_wide=pd.read_csv('~/git/comex_sh_mun/temp/e_c_m_r_pa_clean_wide.csv')
e_wide=e_wide[e_wide.CO_ANO!=2025]
e_wide
```

Funde e salva as tabelas wide de fertilizantes e café.

```python
i_e_wide=i_wide.merge(e_wide)
i_e_wide.to_csv('~/git/comex_sh_mun/temp/i_e_wide.csv',index=False)
i_e_wide
```
