```python
i_f_r_pa=pd.read_csv('~/git/Importacao-por-SH-e-Municipio/temp/i_f_r_pa.csv')
i_f_r_pa.head(5)
```

```python
i_f_r_pa=i_f_r_pa\
.sort_values(by=['Ano','Mês'])\
.rename(columns={'Quilograma Líquido':'KT_f',\
                 'Valor FOB (US$)':'miFOB_f'})
i_f_r_pa[['KT_f','miFOB_f']]=i_f_r_pa[['KT_f','miFOB_f']]/10**6
i_f_r_pa['date']=pd.to_datetime(i_f_r_pa.Ano.astype(str)+'/'+i_f_r_pa.Mês.astype(str))
d_i=i_f_r_pa[['date']].iloc[0][0]
d_f=i_f_r_pa[['date']].iloc[-1][0]
i_f_r_pa[['Ano','Mês']]=i_f_r_pa[['Ano','Mês']].astype('category')
i_f_r_pa=i_f_r_pa.groupby(by=['Ano','Mês'])[['KT_f','miFOB_f']].sum().reset_index()
i_f_r_pa['date']=pd.to_datetime(i_f_r_pa.Ano.astype(str)+'/'+i_f_r_pa['Mês'].astype(str))
i_f_r_pa=i_f_r_pa[(i_f_r_pa.date>=d_i)&(i_f_r_pa.date<=d_f)]
i_f_r_pa
```

```python
e_c_r_pa=pd.read_csv('~/git/Importacao-por-SH-e-Municipio/temp/e_c_r_pa.csv')
e_c_r_pa=e_c_r_pa\
.sort_values(by=['Ano','Mês'])\
.rename(columns={'Quilograma Líquido':'KT_c',\
                 'Valor FOB (US$)':'miFOB_c'})
e_c_r_pa[['KT_c','miFOB_c']]=e_c_r_pa[['KT_c','miFOB_c']]/10**6
e_c_r_pa['date']=pd.to_datetime(e_c_r_pa.Ano.astype(str)+'/'+e_c_r_pa.Mês.astype(str))
d_i=e_c_r_pa[['date']].iloc[0][0]
d_c=e_c_r_pa[['date']].iloc[-1][0]
e_c_r_pa[['Ano','Mês']]=e_c_r_pa[['Ano','Mês']].astype('category')
e_c_r_pa=e_c_r_pa.groupby(by=['Ano','Mês'])[['KT_c','miFOB_c']].sum().reset_index()
e_c_r_pa['date']=pd.to_datetime(e_c_r_pa.Ano.astype(str)+'/'+e_c_r_pa['Mês'].astype(str))
e_c_r_pa=e_c_r_pa[(e_c_r_pa.date>=d_i)&(e_c_r_pa.date<=d_c)]
e_c_r_pa
```

```python
chuvas_r_pa=pd.read_csv('~/git/Importacao-por-SH-e-Municipio/temp/chuvas_r_pa.csv')
chuvas_r_pa=chuvas_r_pa\
.sort_values(by=['Ano','Mês'])


chuvas_r_pa['date']=pd.to_datetime(chuvas_r_pa.Ano.astype(str)+'/'+chuvas_r_pa.Mês.astype(str))

d_i=chuvas_r_pa[['date']].iloc[0][0]
d_f=chuvas_r_pa[['date']].iloc[-1][0]

chuvas_r_pa[['Ano','Mês']]=chuvas_r_pa[['Ano','Mês']].astype('category')
chuvas_r_pa=chuvas_r_pa.groupby(by=['Ano','Mês'])[['mm']].sum().reset_index()
chuvas_r_pa['date']=pd.to_datetime(chuvas_r_pa.Ano.astype(str)+'/'+chuvas_r_pa['Mês'].astype(str))
chuvas_r_pa=chuvas_r_pa[(chuvas_r_pa.date>=d_i)&(chuvas_r_pa.date<=d_f)]
chuvas_r_pa
```

```python
r_pa=i_f_r_pa.merge(e_c_r_pa).merge(chuvas_r_pa)
r_pa=r_pa.rename(columns={'Ano':'Ano1'}).drop(columns='date')
r_pa
```
