```python
r=pd.read_csv('~/git/Importacao-por-SH-e-Municipio/temp/q1.csv')
r
```

```python
col=r.columns[[0,3,4]].tolist()
col
```

```python
best=pd.DataFrame()
for c1 in col:
    a=np.abs(r.groupby(c1)[['test_f%_RFR']].mean()).rename(columns={'test_f%_RFR':'media_erro%'})

    b=np.abs(r.groupby(c1)[['test_f%_RFR']].std()).rename(columns={'test_f%_RFR':'std_erro%'})

    c=round(pd.concat([a,b],axis=1))

    c['erro_max']=c['media_erro%']+c['std_erro%']
    c=c.sort_values(by='erro_max').reset_index(drop=False)
    display(c)
    #display(c.loc[0,c1])
    best=pd.concat([best,c[[c1]].iloc[0]])
best=best.transpose()
display(best)
```

```python
best.to_csv('~/git/Importacao-por-SH-e-Municipio/temp/best.csv',index=False)
```

```python
r_b=r.merge(best)
r_b
```
