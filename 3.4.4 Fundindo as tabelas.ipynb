Fertilizante

```python
i=pd.read_csv('~/git/comex_sh_mun/temp/i_f_m_r_pa_clean.csv')
i=i[i.CO_ANO!=2024]
i=i.groupby(by=['CO_ANO','NO_MUN_i_f','SH4_f','CO_PAIS_e_f'])[['KG_f','VL_f']].sum().reset_index()
i1=i[['CO_ANO','KG_f']]
i2=i.drop(columns='KG_f')
i2.CO_ANO=i2.CO_ANO+1
i=i1.merge(i2,how='outer')
i.to_csv('~/git/comex_sh_mun/temp/i.csv',index=False)
```

Café

```python
e=pd.read_csv('~/git/comex_sh_mun/temp/e_c_m_r_pa_clean.csv')
e=e[e.CO_ANO!=2024]
e=e.groupby(by=['CO_ANO','NO_MUN_e_c','SH4_c','CO_PAIS_i_c'])[['KG_c','VL_c']].sum().reset_index()
e1=e[['CO_ANO','KG_c']]
e2=e.drop(columns='KG_c')
e2.CO_ANO=e2.CO_ANO+1
e=e1.merge(e2,how='outer')
```

Fundindo i + e

```python
i_e=i.merge(e,how='outer')
i_e.to_csv('~/git/comex_sh_mun/temp/i_e.csv',index=False)
i_e
```

```python
c=pd.read_csv('~/git/comex_sh_mun/temp/chuvas_m_r_pa_clean.csv')
c=c[c.CO_ANO!=2024]

c=c.groupby(by=['CO_ANO','Estação']).mm.sum().reset_index()
c1=c[['CO_ANO','Estação','mm']]
c2=c.drop(columns='mm')
c2.CO_ANO=c2.CO_ANO+1
c=c1.merge(c2,how='outer')
c
```

```python
i_e_c=i_e.merge(c)
i_e_c.to_csv('~/git/comex_sh_mun/temp/i_e_c.csv',index=False)
i_e_c
```
